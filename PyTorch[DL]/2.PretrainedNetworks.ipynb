## AlexNet

```python
# To use a pre-trained Alexnet,
# we create an instance of the network
alexnet = models.AlexNet()
```

```python
# create an instance of ResNet with 101 layers
resnet = models.resnet101(pretrained=True)
```

```python
resnet
```

```python
# basis preprocessing step
from torchvision import transforms
```

```python
preprocess = transforms.Compose([
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(
        mean = [0.485, 0.456, 0.406],
        std = [0.229, 0.224, 0.225])
])
```

```python
from PIL import Image
```

```python
img = Image.open("Leo.JPG")
```

```python
img
```

```python
# pass the image through preprocessing pipeline
img_t = preprocess(img)
```

```python
batch_t = torch.unsqueeze(img_t, 0)
```

### Inference : Process of running pre-trained model on new data

```python
# To perform inference, we put the network in eval mode
resnet.eval()
```

```python
out = resnet(batch_t)
```

```python
out
```

```python
with open("imagenet1000_clsidx_to_labels.txt") as f:
    labels = [line.strip() for line in f.readlines()]
```

```python
# extract the maximum value in output tensor using max function
_, index = torch.max(out, 1)
```

```python
# normalize the output tensor
percentage = torch.nn.functional.softmax(out, dim=1)[0]*100
labels[index[0]], percentage[index[0]].item()
```

```python
# arrange percentage values in descending order
_, indices = torch.sort(out, descending=True)
```

```python
[(labels[idx], percentage[idx].item()) for idx in indices[0][:5]]
```
